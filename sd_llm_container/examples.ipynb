# Generative models

```python
import torch
from transformers import pipeline
```

```python
# The model is not included in the container. Make sure to download it separately.
# Also, this model doesn't work for Finnish. For any generative models useful 
# in practice, we probably need GPUs in SD-desktop and a GPU version of the container.
pipe = pipeline(
    "text-generation",
    model="./models/TinyLlama--TinyLlama-1.1B-Chat-v1.0",
    dtype=torch.bfloat16
)
```

```python
# copied from: https://huggingface.co/TinyLlama/TinyLlama-1.1B-Chat-v1.0
messages = [
    {
        "role": "system",
        "content": "You are a friendly chatbot who always responds in the style of a pirate",
    },
    {"role": "user", "content": "How many helicopters can a human eat in one sitting?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])
```

```python
# Test with FinBERT

from bertopic import BERTopic
from bertopic.vectorizers import ClassTfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

#topic_model = BERTopic(embedding_model = document_embeddings, language = "finnish",
#                       nr_topics = 20, calculate_probabilities = True, verbose = True) # calculate_probabilities = True pitäisi laskea jokaisen topicin todennäköisyys topicille

#topics, probs = topic_model.fit_transform(train_docs)

# 14 min with nr_topics = "auto" & calculate_probabilities = True
# 14 min with nr_topics = 20 & calculate_probabilities = True

vectorizer_model = CountVectorizer(stop_words=stopw, ngram_range=(1, 1))
ctfidf_model = ClassTfidfTransformer(reduce_frequent_words=True)
topic_model = BERTopic(embedding_model = "TurkuNLP/sbert-cased-finnish-paraphrase", language = "finnish", vectorizer_model = vectorizer_model, # Muokkaa embedding model
                       umap_model = umap_model, hdbscan_model=hdbscan_model,
                       nr_topics = "auto", calculate_probabilities = False, verbose = True) # Calculate probabilities pois, kunnes toimii muuten järkevästi

topics, probs = topic_model.fit_transform(train_docs)
```
